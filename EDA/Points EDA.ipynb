```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
points = pd.read_csv("points.csv")
```

```python
points.head()
```

```python
points["transfers"].value_counts()
```

```python
points["transfers"].value_counts().plot(kind="bar")
```

```python
points["duration"].describe()
```

```python
plt.figure(figsize=(15,15))
plt.xlabel('Trip Duration')
plt.ylabel('Count of Trips')
plt.title('Histogram of Trip Duration')
plt.grid(True)
plt.hist(points["duration"], alpha=0.75, bins=9, edgecolor='black')
```

```python
points["walkTime"].describe()
```

```python
plt.figure(figsize=(15,15))
plt.xlabel('Walk Time')
plt.ylabel('Count of Trips')
plt.title('Histogram of Walk Time')
plt.grid(True)
plt.hist(points["walkTime"], alpha=0.75, bins=9, color="green", edgecolor='black')
```

```python
points["walkDistance"].describe()
```

```python
plt.figure(figsize=(15,15))
plt.xlabel('Walk Distance')
plt.ylabel('Count of Trips')
plt.title('Histogram of Walk Distance')
plt.grid(True)
plt.hist(points["walkDistance"], alpha=0.75, bins=9, color="orange", edgecolor='black')
```

```python
points["transitTime"].describe()
```

```python
plt.figure(figsize=(15,15))
plt.xlabel('Transit Time')
plt.ylabel('Count of Trips')
plt.title('Histogram of Transit Time')
plt.grid(True)
plt.hist(points["transitTime"], alpha=0.75, bins=9, color="red", edgecolor='black')
```
