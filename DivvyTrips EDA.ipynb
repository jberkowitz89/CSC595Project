```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
divvy_trips["tripduration"].describe()
```

```python
sns.set(style="whitegrid")
ax = sns.boxplot(x=divvy_trips["tripduration"])
```

```python
plt.figure(figsize=(15,15))
plt.xlabel('Trip Duration')
plt.ylabel('Count of Trips')
plt.title('Histogram of Trip Duration')
plt.grid(True)
plt.hist(divvy_trips["tripduration"], alpha=0.75, bins=9, range=(60,5000), edgecolor='black')
```

```python
#Converting start_time and end_time to datetimes
divvy_trips["start_time"] = pd.to_datetime(divvy_trips["start_time"])
```
